```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import networkx as nx
from random import random
import pandas as pd
import seaborn
from collections import defaultdict
```

```python
FILENAME_PATTERN = "../plot_data/cluster_species_%s_%s_%s_u%d.data"

def cluster_graph(inflow_rates,fluc_type,updates=[400000]):
    graphs = {}
    cluster_spop = {}    
    for update in updates:
        graphs[update] = {}
        cluster_spop[update] = {}
        for i in inflow_rates:
            graphs[update][i] = {}
            cluster_spop[update][i] = {}
            for replicate in range(1,31):
            #for replicate in range(1,2):
                G = nx.Graph()
                with open("../data_%s_%s/clustering_history/replicate_%d/%d.output" % (fluc_type, i, replicate, update)) as data:
                    for line in data:
                        line = line.strip().split()
                        if not line[0].startswith("#") and len(line) == 3:
                            G.add_edge(line[0], line[1], length=line[2])

                clusters = nx.components.connected_components(G)
                with open("../data_%s_%s/replicate_%d/detail-%d.spop" % (fluc_type, i, replicate, update)) as data:
                    spop = defaultdict(list)
                    for line in data:
                        items = line.strip().split()
                        for j, cluster in enumerate(clusters):
                            if items and not items[0].startswith("#") and items[0] in cluster:
                                spop[j].append(line)

                for c in spop:
                    with open("../data_%s_%s/clustering_history/replicate_%d/detail-%d_c%d.spop" % (fluc_type, i, replicate, update, c), 'w') as data:
                        data.write('#filetype genotype_data\n')
                        data.write('#format id src src_args parents num_units total_units length merit gest_time fitness gen_born update_born update_deactivated depth hw_type inst_set sequence cells gest_offset lineage\n') 
                        data.write('# Structured Population\n')
                        for line in spop[c]:
                            data.write(line)
                            
                cluster_spop[update][i][replicate] = spop
                graphs[update][i][replicate] = G

    return graphs, cluster_spop

inflow_rates = [1, 10, 100, 1000, 10000]
updates = [100000, 200000, 300000, 400000]
graphs_stag, spop_stag = cluster_graph(inflow_rates, "stag", updates=updates)
graphs_sync, spop_sync = cluster_graph(inflow_rates, "sync", updates=updates)
```

```python
for u, update in graphs_sync.items():
    for i, inflow in update.items():
        #plt.figure(u+i, figsize=(10, 10))
        f, axarr = plt.subplots(5, 6, figsize=(24, 20))
        f.suptitle("stag_%d, update %d" % (i, u))
        f.subplots_adjust(hspace=.5)
        for r, replicate in inflow.items():
            nx.draw(replicate, ax=axarr[(r-1) / 6, (r-1) % 6], with_labels=False)
            axarr[(r-1) / 6, (r-1) % 6].set_title('%d'%r)
        plt.show()
```

```python
NAMES="Not Nand And OrNot Or AndNot Nor Xor Equals".split()

def count_tasks_clusters(inflow_rates, fluc_type, clusters, updates=[400000]):
    stats = {}
    for update in updates:
        stats[update] = {}
        for i in inflow_rates:
            stats[update][i] = defaultdict(list)
            for replicate in clusters[update][i]:
                stats[update][i][replicate] = []
                for c in clusters[update][i][replicate]:
                    filename = "../data_%s_%s/clustering_history/replicate_%d/task_nums-%d_c%d.data" % (fluc_type, i, replicate, update, c)
                    df = pd.read_table(filename, sep=" ", names=NAMES, skiprows=14, index_col=False)
                    stats[update][i][replicate].append(df)
    return stats
```

```python
inflow_rates = [1, 10, 100, 1000, 10000]
updates = [100000, 200000, 300000, 400000]
#tasks_stag = count_tasks_clusters(inflow_rates, "stag", spop_stag, updates=updates)
#tasks_sync = count_tasks_clusters(inflow_rates, "sync", spop_sync, updates=updates)
tasks_stag = count_tasks_clusters([10], "stag", spop_stag, updates=[400000])
tasks_sync = count_tasks_clusters([10], "sync", spop_sync, updates=[400000])
```

```python
def plot_all_replicates(all_data, inflow_rate, update, fluc_type='sync'):
    data = all_data[update][inflow_rate]

    f, axarr = plt.subplots(6, 5, figsize=(20, 30), sharey=True)
    f.suptitle('%s_%d, update %d' % (fluc_type, inflow_rate, update), fontsize=30)
    color = seaborn.color_palette()

    for r in data:
        ax = axarr[(r-1) / 5, (r-1) % 5]
        ax.set_title(str(r))
        for i, t in enumerate(data[r]):
            t.sum().plot(label='Species ' + str(i), ax=ax, kind='bar', color=color[i])
            #t.sum().plot(label='Species ' + str(i), ax=ax)
        ax.legend()

    f.tight_layout()
    f.subplots_adjust(top=0.95)
```

```python
plot_all_replicates(tasks_sync, 10, 400000, fluc_type='sync')
```

```python
plot_all_replicates(tasks_sync, 10, 400000, fluc_type='sync')
```

```python
plot_all_replicates(tasks_stag, 10, 400000, fluc_type='stag')
```

```python
plot_all_replicates(tasks_stag, 10, 400000, fluc_type='stag')
```

```python
def plot_all_heatmaps(all_data, inflow_rate, update, fluc_type='sync'):
    seaborn.set_style('dark')
    data = all_data[update][inflow_rate]
    
    f, axarr = plt.subplots(6, 5, figsize=(50, 50))
    f.suptitle('Resource usage by species\n%s_%d, update %d' % (fluc_type, inflow_rate, update), fontsize=30)
    f.tight_layout()
    f.subplots_adjust(top=0.95)

    for r in data:
        df = pd.DataFrame()
        ax = axarr[(r-1) / 5, (r-1) % 5]
        for i, t in enumerate(data[r]):
            df['Species %d' % i] = t.sum()/len(t)

        ax.matshow(df.T)
        ax.set_xticks(range(9))
        ax.set_title(str(r))

    seaborn.set_style('darkgrid')
```

```python
plot_all_heatmaps(tasks_sync, 10, 400000, fluc_type='sync')
```

```python
plot_all_heatmaps(tasks_stag, 10, 400000, fluc_type='stag')
```
